–ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫: ùëÇ(ùëõ) ‚Äì –∫–æ–ª–∏ –≤–∏—Ö—ñ–¥–Ω–∏–π —Å–ø–∏—Å–æ–∫ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π, —ñ<br>
–∫–æ–∂–µ–Ω –Ω–æ–≤–∏–π –µ–ª–µ–º–µ–Ω—Ç –≤—Å—Ç–∞–≤–ª—è—î—Ç—å—Å—è –±–µ–∑–ø–æ—Å–µ—Ä–µ–¥–Ω—å–æ –∑–∞ –æ—Å—Ç–∞–Ω–Ω—ñ–º –µ–ª–µ–º–µ–Ω—Ç–æ–º.<br>
–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫: ùëÇ(ùëõ2) ‚Äì –∫–æ–ª–∏ –≤–∏—Ö—ñ–¥–Ω–∏–π —Å–ø–∏—Å–æ–∫ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —É<br>
–∑–≤–æ—Ä–æ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É, —ñ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç—É –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø–µ—Ä–µ–±—Ä–∞—Ç–∏ —É—Å—ñ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ<br>
–µ–ª–µ–º–µ–Ω—Ç–∏, —â–æ–± –≤—Å—Ç–∞–≤–∏—Ç–∏ –π–æ–≥–æ –Ω–∞ –ø–æ—Ç—Ä—ñ–±–Ω–µ –º—ñ—Å—Ü–µ<br>
–ê–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –±—É–ª—å–±–∞—à–∫–æ—é<br>
–£ –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É, –∫–æ–ª–∏ –º–∞—Å–∏–≤ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π, –±—É–ª—å–±–∞—à–∫–æ–≤–µ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–æ–∂–µ –∑–∞–≤–µ—Ä—à–∏—Ç–∏<br>
–≤–∏–∫–æ–Ω–∞–Ω–Ω—è –ø—ñ—Å–ª—è –ø–µ—Ä—à–æ–≥–æ –ø—Ä–æ—Ö–æ–¥—É, —è–∫—â–æ –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ–∞–ª—ñ–∑–æ–≤–∞–Ω–∏–π –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é –Ω–∞ –Ω–∞—è–≤–Ω—ñ—Å—Ç—å<br>
–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ (—Ç–æ–±—Ç–æ, —è–∫—â–æ –Ω–∞ –∂–æ–¥–Ω–æ–º—É –ø—Ä–æ—Ö–æ–¥—ñ –Ω–µ –±—É–ª–æ –∑–º—ñ–Ω, –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–∏–ø–∏–Ω—è—î –≤–∏–∫–æ–Ω–∞–Ω–Ω—è). –£ —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ–π —Å–∫–ª–∞–¥–∞—î: ùëÇ(ùëõ)<br>
–£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É (–∫–æ–ª–∏ –º–∞—Å–∏–≤ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —É –∑–≤–æ—Ä–æ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É), <br>
–∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç –Ω–∞ –∫–æ–∂–Ω–æ–º—É –ø—Ä–æ—Ö–æ–¥—ñ –ø–æ—Ç—Ä—ñ–±–Ω–æ –±—É–¥–µ –ø–æ—Ä—ñ–≤–Ω—è—Ç–∏ –∑ –∫–æ–∂–Ω–∏–º —ñ–Ω—à–∏–º –µ–ª–µ–º–µ–Ω—Ç–æ–º.<br>
–û—Ç–∂–µ, —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –≤ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É: ùëÇ(ùëõ2)

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # –í–≤–µ–¥–µ–º–æ –∑–º—ñ–Ω–Ω—É flag –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏, —á–∏ –±—É–ª–∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        swapped = False
        # –ü—Ä–æ—Ö–æ–¥–∏–º–æ –ø–æ –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç–∞—Ö, –∫—Ä—ñ–º –æ—Å—Ç–∞–Ω–Ω—ñ—Ö i –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ (–≤–æ–Ω–∏ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω—ñ)
        for j in range(0, n-i-1):
            # –Ø–∫—â–æ –ø–æ—Ç–æ—á–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –±—ñ–ª—å—à–∏–π –∑–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏–π, –º—ñ–Ω—è—î–º–æ —ó—Ö –º—ñ—Å—Ü—è–º–∏
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # –Ø–∫—â–æ –∑–∞ –ø—Ä–æ—Ö–æ–¥—É –Ω–µ –±—É–ª–æ –∂–æ–¥–Ω–æ—ó –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏, –º–∞—Å–∏–≤ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π
        if not swapped:
            break
    return arr

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
arr = [64, 34, 25, 12, 22, 11, 90]
print("–ù–µ–≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤:", arr)
sorted_arr = bubble_sort(arr)
print("–í—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤:", sorted_arr)
```

```python
import time
import random
import matplotlib.pyplot as plt
# –ê–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥–æ–º –≤—Å—Ç–∞–≤–∫–∏
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# –ê–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥–æ–º –±—É–ª—å–±–∞—à–∫–∏
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –≤–∏–º—ñ—Ä—é–≤–∞–Ω–Ω—è —á–∞—Å—É —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è
def measure_time(sort_function, arr):
    start_time = time.time()
    sort_function(arr)
    return time.time() - start_time

# –ú–∞—Å–∏–≤–∏ —Ä–æ–∑–º—ñ—Ä–æ–º n=[5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]

# –°–ø–∏—Å–∫–∏ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è —á–∞—Å—ñ–≤
times_insert = []
times_bubble = []

# –í–∏–º—ñ—Ä—é–≤–∞–Ω–Ω—è —á–∞—Å—É –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ —Ä–æ–∑–º—ñ—Ä—É –º–∞—Å–∏–≤—É
for n in sizes:
    arr_insert = random.sample(range(1, n*10), n)
    arr_bubble = arr_insert.copy()
    
    # –í–∏–º—ñ—Ä—è—î–º–æ —á–∞—Å –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥–æ–º –≤—Å—Ç–∞–≤–∫–∏
    time_insert = measure_time(insertion_sort, arr_insert)
    times_insert.append(time_insert)
    
    # –í–∏–º—ñ—Ä—è—î–º–æ —á–∞—Å –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥–æ–º –±—É–ª—å–±–∞—à–∫–∏
    time_bubble = measure_time(bubble_sort, arr_bubble)
    times_bubble.append(time_bubble)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_insert, label='Insertion Sort', marker='o', color='b')
plt.plot(sizes, times_bubble, label='Bubble Sort', marker='s', color='r')
plt.xlabel('–†–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É (n)')
plt.ylabel('–ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è (—Å–µ–∫—É–Ω–¥–∏)')
plt.title('–ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è')
plt.legend()
plt.grid(True)
plt.show()
```
