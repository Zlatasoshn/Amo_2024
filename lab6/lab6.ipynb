### **2.Написати процедуру видалення заданої гілки дерева.**

```python
# Реалізація бінарного дерева

def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])

def getRootVal(root):
    return root[0]

def setRootVal(root, newVal):
    root[0] = newVal

def getLeftChild(root):
    return root[1]

def getRightChild(root):
    return root[2]

# Функція для виводу дерева у вигляді текстового дерева

def printTree(tree, level=0):
    if tree:
        printTree(getRightChild(tree), level + 1)
        print(" " * 4 * level + "->", getRootVal(tree))
        printTree(getLeftChild(tree), level + 1)

# Функція для видалення заданої гілки дерева
def deleteBranch(root, value):
    if not root:
        return None
    
    if getRootVal(root) == value:
        return []  # Видалити гілку
    
    leftChild = getLeftChild(root)
    rightChild = getRightChild(root)
    
    if leftChild and getRootVal(leftChild) == value:
        root[1] = []  # Видалити ліву гілку
    else:
        deleteBranch(leftChild, value)
    
    if rightChild and getRootVal(rightChild) == value:
        root[2] = []  # Видалити праву гілку
    else:
        deleteBranch(rightChild, value)
    
    return root

# Приклад створення дерева
tree = BinaryTree(17)
insertLeft(tree, 8)
insertLeft(getLeftChild(tree), 4)
insertRight(getLeftChild(tree), 12)
insertLeft(getLeftChild(getLeftChild(tree)), 2)
insertRight(getLeftChild(getLeftChild(tree)), 6)
insertLeft(getRightChild(getLeftChild(tree)), 10)
insertRight(getRightChild(getLeftChild(tree)), 14)
insertRight(tree, 25)
insertLeft(getRightChild(tree), 20)
insertRight(getRightChild(tree), 30)
insertLeft(getLeftChild(getRightChild(tree)), 18)
insertRight(getLeftChild(getRightChild(tree)), 22)
insertLeft(getRightChild(getRightChild(tree)), 28)
insertRight(getRightChild(getRightChild(tree)), 32)

# Виведення дерева
print("Створене дерево:")
printTree(tree)

# Видалення гілки за заданим значенням
value_to_delete = 25
print(f"\nВидалення гілки за значенням {value_to_delete}...")
deleteBranch(tree, value_to_delete)

# Виведення оновленого дерева
print("\nОновлене дерево:")
printTree(tree)
```

- **Найгірший випадок**: \(O(n)\), якщо дерево перетворюється на "список".

## Delete (Видалення гілки)
- **Середній випадок**: \(O(\log n)\), якщо дерево збалансоване.
- **Найгірший випадок**: \(O(n)\), якщо необхідно видалити гілку в "глибокому" дереві.

### **5.Додати до нього елемент, який дорівнює вашому порядковому номеру у списку групи**
### **6.Вилучити максимальний елемент з купи**

```python
import heapq
import random


def build_max_heap(arr):
    arr = [-x for x in arr]
    heapq.heapify(arr)
    return arr

# Додавання елемента до купи (20)
def insert_to_heap(heap, value):
    # Додаємо елемент, змінюючи його знак на від'ємний
    heapq.heappush(heap, -value)

# Видалення максимального елемента з купи
def delete_max_from_heap(heap):
    # Видаляємо максимальний елемент, змінюючи його знак на від'ємний
    max_value = -heapq.heappop(heap)
    return max_value

random_array = random.sample(range(1, 100), 10)  # Рандомний масив
print("Рандомний масив:", random_array)

max_heap = build_max_heap(random_array)  # Створення купи
print("max купа:", [-x for x in max_heap])  # Перетворюємо значення назад в позитивні

# Додаємо елемент, який дорівнює 20 
insert_to_heap(max_heap, 17)
print("Додаємо елемент 17:", [-x for x in max_heap])

# Видаляємо максимальний елемент
max_element = delete_max_from_heap(max_heap)
print("Видаляємо максимальний елемент:", max_element)
print("Купа після видалення:", [-x for x in max_heap])
```

## **7.Оцінити асимптотичну складність (у середньому і в найгіршому випадку) процедур search, insert і delete роботи з купою.**
## 1. Search
У купі немає прямого способу ефективного пошуку довільного елемента, оскільки вона організована для швидкого доступу лише до кореня. Для загального випадку доведеться обійти всі елементи.

- **Середній випадок**: O(n), де n — кількість елементів.
- **Найгірший випадок**: O(n), якщо шуканий елемент знаходиться в останньому вузлі.

## 2. Insert
Під час вставки нового елемента: елемент додається в кінець купи. Проводиться "просіювання вгору" (heapify up), щоб відновити властивість купи. У гіршому випадку елемент "підніметься" до кореня.

- **Середній випадок**: O(log n), оскільки глибина купи h = log2(n).
- **Найгірший випадок**: O(log n), коли елемент підніметься до самого кореня.

## 3. Delete
При видаленні: кореневий елемент замінюється останнім елементом купи. Останній елемент видаляється. Виконується "просіювання вниз" (heapify down), щоб відновити властивість купи. У гіршому випадку елемент опуститься до листа.

- **Середній випадок**: O(log n).
- **Найгірший випадок**: O(log n), якщо елемент опуститься до самого листа.

## **1.Чим відрізняється структура бінарне дерево він бінарного дерева пошуку?**

- **Бінарне дерево**:
  - Кожен вузол має два нащадки (ліву та праву гілки).
  - Немає обмежень щодо порядку елементів у дереві.
  - Структура може мати різні варіанти розташування елементів.

- **Бінарне дерево пошуку (BST)**:
  - Кожен вузол має два нащадки, причому:
    - Усі елементи в лівому піддереві повинні бути меншими за значення вузла.
    - Усі елементи в правому піддереві повинні бути більшими за значення вузла.
  - Це дозволяє ефективно виконувати операції пошуку, вставки та видалення з асимптотичною складністю O(log n) у середньому випадку.
## **2.Чим відрізняється структура бінарне дерево від бінарної купи?**
- **Бінарне дерево**:
  - Структура з двома піддеревами для кожного вузла, без жорстких правил для значень елементів.
  - Порядок елементів не визначений, що може бути зручним для деяких задач, але не дозволяє оптимізувати операції пошуку.

- **Бінарна купа**:
  - Це спеціалізоване дерево, яке зберігає елементи так, щоб виконувалася одна з властивостей купи:
    - **Максимальна купа**: значення в кожному вузлі повинно бути більшим або рівним за значення його нащадків.
    - **Мінімальна купа**: значення в кожному вузлі повинно бути меншим або рівним за значення його нащадків.
  - Використовується для ефективних операцій вставки та видалення кореня (максимум або мінімум), але не гарантує впорядкованості елементів у всьому дереві.
