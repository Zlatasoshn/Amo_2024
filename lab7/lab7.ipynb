# Контрольні питання

### 1. Що таке граф у термінах теорії графів? Наведіть приклади реальних ситуацій, де можна застосовувати графи.
**Граф** — це математична структура, яка складається з множини вершин (вузлів) і множини ребер (зв'язків), які з'єднують пари вершин. Ребра можуть бути орієнтованими або неорієнтованими.

**Приклади реальних ситуацій**:
- **Соціальні мережі**: користувачі є вершинами, а зв'язки між ними (дружба, підписка) — ребра.
- **Транспортні мережі**: міста або станції — це вершини, а дороги або лінії метро — це ребра.
- **Комп'ютерні мережі**: пристрої (комп'ютери, маршрутизатори) як вершини, а канали зв'язку між ними — ребра.

---

### 2. Які основні види графів існують? Наведіть відмінності між орієнтованими і неорієнтованими графами.
**Основні види графів**:
- **Орієнтований граф (Directed Graph, Digraph)**: у цьому графі ребра мають напрямок, тобто вони представлені впорядкованими парами вершин. Наприклад, зв'язки між веб-сторінками, де одна сторінка може посилатися на іншу.
- **Неорієнтований граф (Undirected Graph)**: у цьому графі ребра не мають напрямку, тобто вони є невпорядкованими парами вершин. Наприклад, мережі друзів у соціальних мережах, де зв'язки є взаємними.

**Відмінності**:
- В орієнтованому графі кожне ребро має початкову і кінцеву вершину (відправлення і призначення).
- В неорієнтованому графі ребро з'єднує дві вершини без визначеного напрямку.

---

### 3. Як можна представити граф у пам’яті комп’ютера? Опишіть структури даних, які використовуються для зберігання графів.
**Графи можна представляти такими структурами даних**:
- **Матриця суміжності**: двовимірна матриця, де елементи вказують на наявність або відсутність ребра між вершинами. Це зручний спосіб для повних графів.
- **Список суміжності**: для кожної вершини зберігається список її сусідів. Це зручніше для розріджених графів, де кількість ребер значно менша за кількість можливих пар вершин.
- **Список ребер**: зберігається набір усіх ребер у графі, що підходить для завдань, де важливі самі ребра.

---

### 4. Як працює алгоритм пошуку в ширину (BFS) на графах? Наведіть приклади ситуацій, де застосовується цей алгоритм.
**Алгоритм BFS** починає з початкової вершини, досліджує всі її сусіди, потім переходить до сусідів сусідів і так далі, обробляючи рівні графа по черзі. BFS використовує чергу для зберігання відвіданих вершин і для забезпечення черговості дослідження.

**Приклади застосування BFS**:
- Пошук найкоротших шляхів у графах (наприклад, на шахівниці або в транспортних мережах).
- Аналіз соціальних мереж для знаходження найближчих зв'язків між людьми або пошуку "друзів друзів".
  
---

### 5. Що таке алгоритм пошуку в глибину (DFS) на графах? Як він відрізняється від BFS? Дайте приклади задач, де використовується DFS.
**Алгоритм DFS** працює за стратегією "глибина перед шириною", тобто він спочатку йде вглиб до кінця шляху, перш ніж повертатися і досліджувати інші можливості. DFS зазвичай реалізується через стек або рекурсію.

**Відмінності між DFS і BFS**:
- BFS обходить граф рівень за рівнем, використовуючи чергу.
- DFS йде по шляху до кінця, повертаючись назад, використовуючи стек.

**Приклад задач, де використовується DFS**:
- **Топологічне сортування** в направлених ациклічних графах.
- **Пошук циклів** у графі.
- **Перевірка зв'язності** в графах.

### 6. Опишіть алгоритм Дейкстри для пошуку найкоротшого шляху в графі. Які умови повинні виконуватися для успішної роботи цього алгоритму?
**Алгоритм Дейкстри** знаходить найкоротший шлях від однієї вершини до всіх інших вершин графа, якщо граф має невід'ємні ваги ребер. Алгоритм працює шляхом поступового розширення найбільш обіцяючих шляхів від початкової вершини, вибираючи найкоротші шляхи за допомогою пріоритетної черги.

**Умови для роботи алгоритму Дейкстри**:
- Граф повинен мати **невід'ємні ваги** на ребрах. Алгоритм не працює з графами, де є від'ємні ваги, оскільки це може призвести до некоректних результатів.


# Завдання для самостійної роботи
## 1.Для варіанту, заданого викладачем, задати і візуалізувати граф за допомогою коду, наведеного у цій роботі.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Створення графа
G = nx.Graph()

# Додавання вершин
nodes = ['1', '2', '3', '4', '5', '6', '7']
G.add_nodes_from(nodes)

# Додавання ребер з вагами
edges = [
    ('1', '4', 14), 
    ('1', '2', 9), 
    ('2', '3', 13), 
    ('3', '7', 10), 
    ('7', '5', 11), 
    ('7', '1', 6),
    ('5', '6', 8), 
    ('6', '7', 7), 
    ('4', '7', 12), 
    ('2', '5', 10)
]

G.add_weighted_edges_from(edges)

# Позиції для розміщення вершин
pos = nx.spring_layout(G, seed=42)  # Встановлення seed для стабільного розташування

# Візуалізація графа
plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(G, pos, node_size=800, node_color='lightgreen')
nx.draw_networkx_labels(G, pos, font_size=16, font_family='sans-serif', font_weight='bold')
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color='darkgray')

# Виведення ваг ребер
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14)

# Показати графік
plt.title("Граф з 8 вершинами і 10 ребрами з вагами", fontsize=18)
plt.axis('off')
plt.show()

```

## 2. Для свого варіанту реалізувати всі алгоритми, згідно з прикладами, наведеними вище.
### Алгоритм Дейкстри

```python
# Імпорт бібліотек
import networkx as nx

# Алгоритм Дейкстри
length, path = nx.single_source_dijkstra(G,source='1')

# Виведення довжин найкоротших шляхів
print("Довжини найкоротших шляхів з вершини 1:")
print(length)

# Виведення самих шляхів
print("\nШляхи:")
print(path)
```

### Алгоритм Беллмана-Форда

```python
# Алгоритм Беллмана-Форда
length_bf = nx.single_source_bellman_ford_path_length(G, source='1')

# Виведення довжин найкоротших шляхів
print("Довжини найкоротших шляхів з вершини 1 (Беллман-Форд):")
print(length_bf)
```

### DFS

```python
# Пошук в глибину для знаходження всіх шляхів
def dfs_paths(graph, start, goal, path=None):
    if path is None:
        path = [start]
    if start == goal:
        yield path
    for next in set(graph.adj[start].keys()) - set(path):
        yield from dfs_paths(graph, next, goal, path + [next])

print("Всі шляхи від 1 до 5 (DFS):")
all_paths = list(dfs_paths(G, '1', '5'))
for p in all_paths:
    print(p)
```

### BFS

```python
# Пошук в ширину для знаходження найкоротших шляхів
def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        vertex, path = queue.pop(0)
        for next in set(graph.adj[vertex].keys()) - set(path):
            if next == goal:
                yield path + [next]
            else:
                queue.append((next, path + [next]))

print("Всі шляхи від 1 до 5 (BFS):")
bfs_result = list(bfs_paths(G, '1', '5'))
for p in bfs_result:
    print(p)
```
